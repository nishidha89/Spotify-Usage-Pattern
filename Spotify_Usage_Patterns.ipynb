### Exploratory Data Analysis


```python
#Checking unique values in age column
dataset['Age'].unique()
```

```python
#Checking unique values in Gender column
dataset["Gender"].unique()
```

### 1. What is the predominant gender demographic among Spotify users?

```python
plt.pie(dataset['Gender'].value_counts(),autopct="%1.1f%%",labels=['Female','Male','Others'])
plt.title("Gender distribution of Spotify users")
plt.legend(loc="upper right")
plt.show()
```

### Most of Females use Spotify which are 75%

### 2. Which age group people mostly use Spotify?

```python
labels=['20-35','12-20','6-10','35-60','60+']
plt.pie(dataset['Age'].value_counts(),labels=labels,autopct="%1.1f%%")
plt.title("Age distribution of Spotify users")
plt.legend(loc="upper right")
plt.show()
```

### 81% of users who use Spotify falls under the age group of 20-35

### 3. Among the age group in which most of the users falls to which gender demography do these users belong?

```python
sns.countplot(data=dataset,x="Age",hue="Gender")
```

### As most of the Spotify users falls under age group of 20-35 among them female users are more
